## **a. Basit Doğrusal Regresyon**
* Temel amaç, bağımlı ve bağımsız değişken arasındaki ilişkiyi ifade eden doğrusal fonksiyonu bulmaktır.
* Basit demek bir adet bağımsız değişkenden oluşan demektir. Dolayısıyla bir adet bağımlı değişken ve bir adet bağımsız değişken modelimizde olacaktır.
* Modellemeyi gerçekleştirdiğimizde elimizde olusan denklem asagidaki denklem olacaktir.
    * $\hat{y_i} = b_0 + b_1x_i$ 
    * $x_i$ bagimsiz degisken degerlerini ifade eder.
    * $b_0, b_1$ veriseti icerisinden bulmamiz gereken parametrelerdir.
    * $\hat{y_i}$ ise tahmin edilen degerleri ifade etmektedir.
* Verimizin dogrusal bir model ile temsilini karar verdikten sonra asagidaki tanimlamalar yapilir.
    * <div style="text-align:center"> </div> 
* Matematiksel olarak $b_0, b_1$ degerlerinin hesaplanmasi ise asagidaki adimlarin gerceklestirilmesi ile formülize edilebilir.
    * <div style="text-align:center"> </div> 
    * <div style="text-align:center"> </div> 
    * <div style="text-align:center"> </div> 
    * Bu asamadan sonra b0 ve b1'e gore turev alindiginda elde edilen sonuclar asagidaki gibi formulize edilir.
    * <div style="text-align:center"> </div> 
    * <div style="text-align:center"> </div>

## **b. Çoklu Doğrusal Regresyon**

Basit doğrusal regresyondan farkı birden fazla bağımsız değişkenin bağımlı değişkeni etkileme durumudur. Genel formulasyonu aşağıdaki gibidir.
* <div style="text-align:center"> </div> 
* Buradaki $\beta$ degerleri icin model icindeki bagimsiz degiskenlerin etkilerini kontrol etmek icin kullanilir. Bu degerleri bulmak icin de asagidaki denklemler cozulur.
* Hata kareler toplami formulundeki esitlikten $\beta$ degerin formulunu cikartiriz. <div style="text-align:center"> </div>
* Cikartilan bu formulasyon <div style="text-align:center"> </div>
* Burada X degeri bagimsiz degiskenlerden olusan matrisi ifade etmektedir. Y degeri de bagimli degiskeni ifade etmektedir. Bu islemin sonucunda bir vektor(array) ortaya cikacaktir.
* Eger gozlem sayisi n>1000 gibi bir sinira ulasilmissa formulasyondaki tersine alma islemi imkansiza dogru yaklasacagi icin bu parametreleri bulmak adina cesitli yontemler de kullanilir. Buna bir ornek olarak *Gradient Descent* yontemini gosterebiliriz.

### Doğrusal Regresyonun Varsayımları
Eger kullanmis oldugumuz dogrusal regresyon modelini uzerine is kararlari alinacak, dogrusal neden sonuc baglaminda bazi yorumlamalar yapilacak seklinde dusunuyorsak bu baglamda varsayimlara dikkat etmeliyiz. Cunku yapilacak olan cikarimlar, genellemeler ve neden-sonuc yorumlamalari ancak ve ancak bu varsayimlar saglandiginda gerceklestirilebilmektedir.
* Hatalar normal dagilir.
* Hatalar birbirinden bagimsizdir ve aralarinda otokorelasyon yoktur.
* Her bir gozlem icin hata terimleri varyanslari sabittir.
* Degiskenler ile hata terimleri arasinda iliski yoktur.
* Bagimsiz degiskenler arasinda coklu dogrusal iliski problemi yoktur.

### Regresyon Modelinin Avantajları ve Dezavantajları
* (+) Iyi anlasilirsa diger tum ML ve DL konulari cok daha rahat kavranir.
* (+) Dogrusallik ve nedensellik yorumlarinin yapilabilmesini saglar, bu durum aksiyoner ve stratejik olarak modelleme imkani verir.
* (+) Degiskenlerin etki duzeyleri ve anlamliliklari degerlendirilebilir.
* (+) Bagimli degiskendeki degiskenligin aciklanma basarisi olculebilir.
* (+) Model anlamliligi degerlendirilebilir.
* (-) Varsayimlari bulunmaktadir.
* (-) Aykiri gozlemlere fazla duyarlidir.

### MLR - Model Oluşturma

```python
import pandas as pd
df = pd.read_csv("./csv/Advertising.csv")
Y = df[["Sales"]]
X = df.drop(["Unnamed: 0","Sales"],axis= 1)
```

Model ile alakali istatistiksel detaylara ulasmak icin *statsmodel.api* altindaki *OLS* metodunu kullaniyoruz.

```python
from statsmodels.api import OLS
lm = OLS(Y,X)
model = lm.fit()
```

Burada modelimizin ozeti yer almaktadir.
* $R^2$ degeri, bagimsiz degiskenlerin bagimli degiskendeki degisikligi aciklama yuzdesidir.
* $Adjusted R^2$ degeri ise $R^2$ degerinin duzeltilmis halidir. $R^2$ degeri, verisetine ne kadar degisken eklenirse o kadar sismeye megillidir. Bu durum yorumlama icin yaniltici olmaktadir, bunun onune gecmek icin $Adjusted R^2$ degeri olusturulmustur. Bu da daha saglikli yorum yapmamiza olanak tanimaktadir.
* $F-statistic$ degeri, modelin anlamliligini test etmek icin kullanilan bir istatistiktir.
* $F-statistic$'nin P-Value degeri burada $Prob(F-statistic)$ olarak verilmistir. Bu degerin 0.05'ten kucuk olmasi anlamli oldugu sonucu bize vermektedir.
<br><br>
Asagidaki tablo icin;
* coef ifadesi, her degiskenin katsayisini $\beta$ degerini ifade etmektedir.
* $std err$ degeri standart hatayi temsil etmektedir.
* Degiskenlerin anlamliligini $t$ ve onun P-Value degeri olan $P>|t|$ degerleriyle yorumlariz.

```python
model.summary()
```

Standart saglamak icin modellerimizi scikit-learn uzerinden olusturmaya ve tahmin yapmaya devam edecegiz. Burada Basit Dogrusal Regresyon ve Coklu Dogrusal Regresyon ayni fonksiyon uzerinden calistirilmaktadir.

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(X,Y)
print(model.intercept_)
model.coef_
```

Modelimizde tahmin yapmak icin sirasiyla TV,Radio ve Newspapre icin kendi belirledigimiz degiskenler ile bir DF olusturalim.

```python
df = pd.DataFrame([[122],[15],[5]]).T
df
```

```python
model.predict(df)
```

Simdi modelimizin basarisini olcmek icin kullanilan **Hata Kareler Ortalamasi(MSE)** ni scikit-learn araciligiyla bulalim. Bunun icin **from sklearn.metrics import mean_squared_error** seklinde fonksiyonumuzu dahil ediyoruz. Ilk parametre olarak *gercek degerler*, ikinci parametre olarak ise *tahmin edilen degerleri* girmeliyiz.

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y,model.predict(X))
```

**Hata Kareler Ortalamasinin Karekoku(RMSE)** nu ise karekok islemini gerceklestirerek buluruz.

```python
import numpy as np
np.sqrt(mean_squared_error(Y,model.predict(X)))
```

### MLR - Model Tuning
Bu baslik altinda *Holdout* ve *K-Fold Cross Validation* islemlerini gerceklestirecegiz.

#### i. Holdout Yontemi
Bu asamada verimizi train ve test olarak iki parcaya ayiriyoruz.
* Bu islemi **sklearn.model_selection** altindaki **train_test_split** metodu ile gerceklestiriyoruz.
* Gerceklestirirken sirasiyla X_train,X_test,Y_train,Y_test degerlerini esitligin sol tarafinda belirtmemiz gerekiyor.
* Parametre olarak sirasiyla
    * Bagimsiz Degisken - X
    * Bagimli Degisken - Y
    * Test Verisinin sayisini belirten test_size
    * Her calismada farkli veri setleri olusturmasin diye rastgele bir deger verilen random_state degeri metoda parametre olarak verilir.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
model = lm.fit(X_train,Y_train)
```

Modelimizi egittikten sonra *Egitim Hatasi* ve *Test Hatasi* degerlerini bulacagiz.

```python
RMSE = np.sqrt(mean_squared_error(Y_train,model.predict(X_train)))
print("Egitim Hatasi : ",RMSE)
```

```python
RMSE = np.sqrt(mean_squared_error(Y_test,model.predict(X_test)))
print("Test Hatasi : ",RMSE)
```

#### ii. K Fold Cross Validation
* Bu islemi **sklearn.model_selection** altindaki **cross_val_score** metodu ile gerceklestiriyoruz.
* Parametre olarak sirasiyla
    * Olusturdugumuz model degiskeni
    * Bagimsiz Degisken - X
    * Bagimli Degisken - Y
    * Kac katli bir gerceklestirim istedigimizi belirttigimiz cv degeri
    * Donut olarak nasil bir skorlama istedigimizi belirttigimiz scoring argumani var. Burada istersek r2 istersek de hata kareleri ortalamasini belirtebiliriz. Hata kareleri ortalamasi icin "neg_mean_squared_error" olarak arguman veririz.
* Burada sonuc olarak 10 adet hata kareler ortalamasi degeri donduruldu. Bunun sebebi veriseti 10 parcaya ayrildi her parcada bir adet birim disarda birakilarak onunla birlikte model test edildi. Bu on kez tekrarlandi, her bir deger de her denemenin hata kareler ortalamasinin sonucudur. Fonksiyon geregi sonuclar negatif geldigi icin onune bir - koyarak sonuclari + hale getirdik

```python
from sklearn.model_selection import cross_val_score
-cross_val_score(model,X_train,Y_train,cv=10,scoring="neg_mean_squared_error")
```

**Validasyon islemi gerceklestikten sonraki Egitim Hatasi;**

```python
np.sqrt(np.mean(-cross_val_score(model,X_train,Y_train,cv=10,scoring="neg_mean_squared_error")))
```

**Validasyon islemi gerceklestikten sonraki Test Hatasi;**

```python
np.sqrt(np.mean(-cross_val_score(model,X_test,Y_test,cv=10,scoring="neg_mean_squared_error")))
```

## **c. PCR(Principal Component Reg)- Temel Bileşen Regresyonu**

Değişkenlere boyut indirgeme uygulandıktan sonra çıkan bileşenlere regresyon modeli kurulması fikrine dayanır.
* <div> </div>
* Birçok gerçek hayat probleminde üzerinde çalışıyor olduğumuz verisetinin bağımsız değişkenleribirbirleri ile yüksek korelasyona sahip olabiliyor. Bu da benzer tahminsel bilgileri barındırdıkları anlamına geliyor. Değişkenler arasındaki yüksek korelasyon olması **çoklu doğrusal bağlantı problemi**ne sebep oluyor. Bu problem EKK(En Küçük Kareler) yöntemindeki katsayılarının varyansını arttırarak yanlılığa sebep olmaktadır.
* Gerçek hayat verisetlerinde bazen değişken sayısının gözlem sayısından fazla olması durumu ile karşılaşılabiliyor. Bu da **çok boyutluluk laneti** olarak isimlendiriliyor. 
* Bu durumlardan dolayı varsayımlar sağlanmadığı için çeşitli alternatifler gerçekleştirilmiştir. PCR da bunlardan bir tanesidir.
* PCA(Temel Bileşen Analizi) : Elimizde p adet değişken olduğunda, bu p adet değişkenin içerdiği bilginin yüksek bir kısmını ondan daha az sayıda bir bileşen ile ifade etmek fikrine dayanır. 100 değişkenin içerdiği bilgiyi ondan daha az sayıda değişken ile bu bilginin maksimumunu temsil etmeye çalışmaktır.
* Verisetinde değişken indirmesi yapıldıktan sonra yapılan yeni verisetine regresyon uygulama işlemine PCR denir.

Veriseti beyzbol oyuncularina iliskin bir veri seti. Biz beyzbol oyuncularinin ozelliklerine gore maaslarini degerlendiren bir model kuracagiz.

```python
df = pd.read_csv("./csv/Hitters.csv")
df = df.dropna()
df = df.iloc[:,1:]
dms = pd.get_dummies(df[["League","Division","NewLeague"]])

y = df["Salary"]

X_ = df.drop(["Salary","League","Division","NewLeague"],axis=1).astype("float64")
X = pd.concat([X_,dms[["League_N","Division_W","NewLeague_N"]]],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.25,
                                                   random_state = 0)
```

Scikit-learn içerisinde R'da olduğu gibi PCR'ın direkt olarak uygulamasını barındırmadığı için bunu kendimiz gerçekleştirmemiz gerekiyor. Bu sebeple öncelikle bir PCA işlemi gerçekleştirmemiz gerekiyor, daha sonra oluşturduğumuz PCA üzerine bir regresyon modeli oluşturarak PCR uygulamasını gerçekleştirmiş olacağız.
* PCA nesnesini **sklearn.decomposition** altında **PCA** olarak çağırıyoruz.
* PCA constructorı içerisine verilen *bileşen sayısı argümanı*nı vermiyoruz. Böylece değişken sayısı kadar bileşen oluşmuş olacaktır.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
pca = PCA()
```

* PCA icerisine egitim verisini verirken **scale** isleminin yapilmasi gerekiyor. Bu sebeple **sklearn.preprocessing** altinda **scale** olarak projemize dahil ederiz.
* Bu asamadan sonra elimizde degisken sayisi kadar bilesen olusturulmus fakat aralarindaki baglanti problemi kalkmis, indirgenmis bir veriseti bulunmaktadir.

```python
X_reduced_train = pca.fit_transform(scale(X_train))
```

Ilk 10 bilesenlerce kumulatif olarak aciklanan varyansi gormek istersek;

```python
np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
```

Olusan sonucu yorumlarsak;
* Birinci bilesenin verisetinde bulunan toplam degiskenligin(varyansin) %37.35'ini acikladigi goruluyor.
* Ikinci bilesenin , *birinci bilesen ile birlikte* acikladigi varyans ise %59.36'dir.
* Ozet olarak; bilesen sayisi arttikca orijinal verisetinin icindeki varyansi aciklama oranlarinin arttigi goruluyor.
* Yani burada 19 degiskenle verisetini ifade etmek yerine 8 degisken ile %95 verisetimizi aciklayabiliriz. Indirgeme islemini de bu sekilde yapabiliriz.

Tahmin islemini gerceklestirirken 19 olan degisken sayisini indirgeme islemi yapmadan gerceklestiregiz. Fakat modelimizi indirgenmis olan X degerleri uzerinden egitecegiz. Model Tuning basligi altinda degisken indirgeme islemini gerceklestirecegiz.

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
pcr_model = lm.fit(X_reduced_train,y_train)

y_pred = pcr_model.predict(X_reduced_train)

# RMSE Egitim Hatasi
RMSE = np.sqrt(mean_squared_error(y_train,y_pred))
RMSE
```

```python
pca2 = PCA()
X_reduced_test = pca.fit_transform(scale(X_test))

y_pred = pcr_model.predict(X_reduced_test)

# RMSE Test Hatasi
RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
RMSE
```

### PCR - Model Tuning
Buraya kadar olan kisimda PCA ile indirgenen tum bilesenleri kullanarak hesaplamalar yaptik. Fakat farkli bilesen sayilariyla farkli sonuclar alinabilir.
* PCR'in optimize edilmesi gereken hiperparametresi **buradaki bilesen sayisidir.**
* Asagida manuel olarak train setinde birinci bileseni sectik. Tahmin yaparken de test setini ayni ayara getirmemiz gerekiyor. Yoksa boyut hatasi aliriz.

```python
lm = LinearRegression()
pcr_model = lm.fit(X_reduced_train[:,0:1],y_train)
y_pred = pcr_model.predict(X_reduced_test[:,0:1])
# RMSE Test Hatasi
RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
RMSE
```

Bu hiperparametre ayarlamasini el yordamiyla degil, CV yontemiyle olasi bilesen sayi degerlerini bir vektor olarak gonderip en iyi sonucu veren sayiyi optimum olarak belirleyerek hedefimize ulasacagiz.

* K Fold Cross Validation islemi icin model_selection altindan KFold fonksiyonunu cagiriyoruz.
* KFold fonksiyonuna parametre olarak
    * n_splits = K Fold CV isleminin kac katli olacagini belirtiyoruz.
    * shuffle = Verilerin gruplara ayrilmadan once karistirilip karistilirmayacagini belirtir.
    * random_state = Rastgele olusan degerlerin ayni state degerinde sabit olusmasini saglayan parametredir.
* Asagida cv10 icerisinde 10 katli CV nesnesi olusturuldu.

```python
from sklearn import model_selection
cv10 = model_selection.KFold(n_splits=10,
                            shuffle=True,
                            random_state=1)
```

Her bir bilesen sayisi icin olusturulacak modeller ile CV'na gore Train Hatasi hesaplanacak ve bir kaydetme islemi gerceklestirilecek. Bunun sonunda biz bu sonuclari gorsellestirip degerlendirip (Brute Force) PCR icin en uygun bilesen sayisina karar verecegiz.

## **d. PLS(Partial Least Squares) - Kısmi En Kucuk Kareler Regresyonu**

Degiskenlerin daha az sayida ve aralarinda coklu dogrusal problemi olmayan bilesenlere indirgenip regresyon modeli kurulmasi fikrine dayanir.
* PCR ile ayni mantiktan yuruyor.
* Cok boyutluluk lanetine (p > n) cozum sunar.
* Coklu dogrusal baglanti problemine cozum sunar.
* PLS'de PCR gibi bagimsiz degiskenlerin dogrusal kombinasyonlarini bulur. Bu dogrusal kombinasyonlar bilesen ya da latent(gizil) degisken olarak adlandirilir.
* PLS, NIPALS adli algoritmanin ozel bir halidir. Iteratif olarak bagimli degisken ile yuksek korelasyona sahip degiskenler arasindaki gizil(latent) iliskiyi bulmaya calisir. **PLS ile PCR'i ayiran durumdur.**
* Gorselde **Predictor** olarak gecen tahminciler burada bagimsiz degiskenlerimizdir. Bu bagimsiz degiskenlerin bir kombinasyonu uzerinden bagimli degiskenin tahmini gerceklesiyor.
* Boyut indirgeme yaklasiminda P adet degiskenin tasidigi bildigi ondan daha az sayida olan C adet bilesen ile ifade etmek amaclanmistir.
* <div> </div>

#### PCR vs PLS
* PCR'da dogrusal kombinasyonlar yani bilesenler **bagimsiz degisken uzagindaki degiskenligi** maksimum sekilde ozetleyecek bicimde olusturulur.
* Bu durum bagimli degiskeni aciklama yetenegi olmamasina sebep olmaktadir.
* PLS'te ise **bilesenler bagimli degisken ile olan kovaryansi** maksimum sekidle ozetleyecek bicimde olusturulur.
* *Degiskenler atilmak istenmiyorsa ve aciklanabilirlik araniyorsa: PLS*
* PLS, *gozetimli boyut indigeme proseduru*, PCR *gozetimsiz boyut indirgeme proseduru* olarak goruleblir.
* Iki yontemin de bir tuning parametresi vardir o da **bilesen sayisidir.**
* Optumum bilesen sayisini belirlemek icin **Cross Validation** yontemi kullanilir.

## **e. Ridge Regresyon**

Amaç hata kareler toplamını minimize eden katsayıları, bu katsayılara bir ceza uygulayarak bulmaktır.
* Standart regresyonda  Hata Kareleri Toplamını(EKK) aşağıdaki formül ile bulabiliyoruz. <div> </div>
* Ridge Regresyon içinse <div> </div>
* Ridge,Lasso ve ElasticNet Regresyon, literaturde duzenlilestirme yontemleri olarak gecmektedir. Buradaki $L_2$ ifadesi, buradaki duzenlilestirme durumunu ifade eden normdur. Bu normun ifade ettigi sey ise $\beta_j$ degerinin kare alma islemidir.
* **Özellikleri;**
    * Aşırı öğrenmeye karşı dirençlidir.
    * Yanlıdır fakat varyansı düşüktür. (Bazı durumlarda yanlı modelleri daha çok tercih ederiz)
    * Çok fazla parametre olduğunda klasik regresyona göre daha iyidir.
    * Curse of Dimensionality'e karşı çözüm sunar.
        * Curse of Dimensionality : Değişken sayısının gözlem sayısından fazla olma durumudur.
    * Multicollinearity(Çoklu Doğrusal Bağlantı Sorunu) olduğunda etkilidir.
        * Multicollinearity : Bağımsız değişkenler arasında yüksek korelasyon olması demektir. Yani bir değişkenin taşıdığı bilginin aynısını bir başka değişken de taşımaktadır.
    * Tüm değişkenler ile model kurar, ilgisiz değişkenleri modelden çıkarmaz, katsayılarını sıfıra yaklaştırır.
    * $\lambda$ kritik roldedir. Formüldeki iki terimin göreceli etkilerini kontrol etmeyi sağlar. Ayar parametremizdir. $\lambda$ Ridge regresyonun optimize edilmesi gereken parametresidir.
    * $\lambda$ için iyi bir değer bulunması önemlidir. Bunun için Cross Validation işlemi kullanılır.
* **$\lambda$ Ayar Parametresinin belirlenmesi**
    * $\lambda$ = Ayar parametresidir. Yani kullanıcı tarafından belirlenen, ayarlanan, optimize edilmesi gereken parametredir.
    * $\beta$'lar ise veriseti içerisinden bulunacak olan parametrelerdir.
    * $\lambda$ = 0 olduğunda formül klasik regresyona dönmektedir. Bizim arayışımız ise Hata Kareler Toplamını(SSE) minimum yapan $\lambda$ değeridir.
    * $\lambda$ için belirli değerleri içeren bir küme seçilir ve her birisi için cross validation test hatası hesaplanır.
    * En küçük cross validation'ı veren $\lambda$ ayar parametresi olarak seçilir.
    * Son olarak seçilen bu $\lambda$ ile model yeniden tüm gözlemlere fit edilir.

    * Ridge ve Lasso yontemleri birbirinden ustun degildir.

    * $\lambda$'nin sifir oldugu yer EKK'dir. HKT'yi minimum yapan $\lambda$'yi ariyoruz.

Optimize olan lambda degerini bulduktan sonra final modelimizi olusturup test hatasini hesapliyoruz.

```python
enet_tuned = ElasticNet(alpha=enet_cv_model.alpha_).fit(X_train,y_train)
print("Tuned ElasticNet Test Hatasi:\t",np.sqrt(mean_squared_error(y_test,enet_tuned.predict(X_test))))
```

### Onemli Not:
* "alphas" parametresi icin bir liste olusturmadik ve bunu modele biraktik. ElasticNetCV fonksiyonu uygun olan degeri buldu. Eger olusturulmak istenirse alphas = [.,.,.] seklinde bir liste olusturulabilir ve ElasticNetCV icerisine arguman olarak ifade edilebilir.
* "l1_ratio" parametresi. Bu parametre degeri 0 oldugunda L2 cezelandirmasi, 1 oldugunda ise L1 cezalandirmasi yapar. Dolayisiyla 0-1 arasinda degisimi aslinda cezalandirma metodlarinin goreceli etkilerini ifade eder. Uygulamamizda ikisininde etkisinin ayni duzeyde olmasini istedigimizden dolayi bunun on tanimli degerini oldugu gibi biraktik: 0.5
* "l1_ratio" ve "alphas" beraber dusunuluerse "l1_ratio" icin aranacak bir liste verildiginde yani ceza turlerinin etkilerini ayarlamak icin bir liste verildiginde her bir "l1_ratio" degeri icin farkli "alphas" degerleri olacaktir ki bu durumda bu alphalar ceza terimleri etkilerini modele yansitmis olacak.
