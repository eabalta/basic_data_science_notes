# Makine Öğrenmesi

Bilgisayarların insanlara benzer bir şekilde öğrenmesini sağlamak amacıyla çeşitli algoritma ve tekniklerin geliştirilmesi için çalışılan bilimsel çalışma alanıdır.

## Temel Kavramlar

### Bağımlı & Bağımsız Değişken
* Bağımlı Değişken:
    * Hedef değişken, yanıt değişkeni, çıktı değişkeni gibi isimler de verilir.
    * Makine öğrenmesi probleminde tahmin etmek için hedeflediğimiz ana değişkenimizdir.
    * Doğrusal fonksiyon içerisindeki y değişkenidir.
* Bağımsız Değişken:
    * Bağımlı değişkeni oluşturan diğer değerler ise bağımsız değişkenlerdir.
    * Doğrusal fonksiyon içerisindeki x değişkenleridir.

### Öğrenme Türleri
* **Denetimli Öğrenme:**
    * Bağımlı değişken ve bağımlı değişkeni meydana getiren bağımsız değişkenler çalışma alanında bir aradaysa bu duruma denetimli öğrenme denir.
    * <div> </div> 
* **Denetimsiz Öğrenme:**
    * Bağımlı değişkenin çalışma içinde olmadığı öğrenme türüdür.
    * Bağımsız değişkenlerin belirli eylemler sonucunda bir şeye sebep olması bir yapıyı ortaya çıkarması gibi işaretleme yapılamamaktadır, bir etiketleme yoktur.
    * Sadece gözlem birimlerinden oluşur.
    * Amacımız gözlem birimlerini birbirine benzer özelliklerine göre bir araya getirmektir.
    * <div> </div> 
* **Pekiştirmeli Öğrenme**
    * Pekiştirmeli öğrenme, amaca yönelik ne yapılması gerektiğini öğrenen bir makine öğrenmesi yaklaşımıdır. Pekiştirmeli öğrenmede ajan (agent) adı verilen öğrenen makinemiz karşılaştığı durumlara bir tepki verir ve bunun karşılığında da sayısal bir ödül sinyali alır. Ajan/öğrenen makine aldığı bu ödül puanını maksimuma çıkartmak için çalışır. Bu şekilde çalışan deneme yanılma yöntemi, pekiştirmeli öğrenmenin en ayırt edici özelliğidir.
    * <div> </div> 
* **Yarı Denetimli Öğrenme**
    * Yarı-denetimli öğrenme, denetimli öğrenme ile denetimsiz öğrenme arasında bir yerdedir. Denetimli öğrenme, doğru cevaplarla (hedef değerler) etiketlenmiş eğitim verileriyle başlar. Öğrenme sürecinden sonra, önceden etiketlenmemiş benzer veriler için cevapları tahmin edebilecek şekilde ayarlanmış bir ağırlık (weight) setine sahip bir model kurarız. Yarı-Denetimli öğrenme, bir modeli fit etmek için hem etiketli hem de etiketsiz verileri kullanır. Bazı durumlarda, etiketsiz verilerin eklenmesi modelin doğruluğunu artırır. Diğer durumlarda, etiketlenmemiş veriler modeli daha da kötüleştirebilir; farklı algoritmaların farklı veri özelliklerine karşı güvenlik açıkları vardır.
    * <div> </div>

* Train : Eğitim setimizi temsil etmektedir. Verinin büyük çoğunluğu eğitmek için kullanılır.
* Test : Test setimizi temsil etmektedir. Verinin belirli bir kısmını içermektedir.

## Model Doğrulama Yöntemleri - Model Validation
Model kurmak, bagimli degisken ile bagimsiz degiskenler arasindaki iliskiyi modellemek demektir.<br>
Model dogrulama ise modellerin urettigi sonuuclarin dogru degerlendirilmesi calismalaridir.
* **Holdout Yontemi**
    * Sinama seti yaklasimi olarak da isimlendirilir.
    * Daha once bahsettigimiz test-train yapisiyla birebir ayni mantiktir.
    * <div> </div> 
* **k Katli Capraz Dogrulama - k Fold Cross Validation**
    * Veriseti k adet parcaya ayrilir, daha sonra belirlenen alt kumelerden birisi disarida birakilir. Elde kalan diger kumelerle model olusturulur. Model disarida birakilan kumeyle test edilir. Ayni islem butun parcalar icin tekrar edilir.
    * Bu islemler sonucunda elde edilen hatalarin ortalamasi alindiginda validasyon hatasi hesaplanmis olur.
    * Daha sonra calismanin en basinda bolmus oldugumuz test setiyle de tekrar test ederiz.
    * k fold cross validation islemini test ve train olarak boldugumuz verisetimizin **train** seti uzerinde gerceklestiririz. 
    * <div> </div> 
* **Leave One Out**
    * k fold cross validation isleminin ozel bir halidir. Burada n eleman sayisi kadar kume olusturulur. Her bir kumede sirasiyla bir eleman cikartilarak n kadar alt kumeler olusturulur.
    * Cikarilan gozlem haric kalan gozlemlerle model egitlir, cikarilan gozlem ile ise model test edilir.
    * Teorik olarak guzel fakat verisetinin buyuklugu arttikca uygulanmasi zorlasmaktadir.

## Bias-Variance Tradeoff - YanlılıkVaryans
Model kurarken, modelim başarısını değerlendirirken elimizde iki tip hata olur.
* **Eğitim Hatası:**
    * Modeli kurmak için kullanılan veriseti üzerinden elde ettiğimiz hatalardır.
* **Test Hatası:**
    * Kurulan modeli test etmek için kullandığımız veriseti için oluşan hatadır.
* *Eğitim hatası test hatasının yanlı ve kötü bir tahmincisidir.* Yani birbirlerine bağımlılıkları bulunmaktadır.
* **Esneklik**, verinin fonksiyonel yapısının uygun bir şekilde yorumlanmasıdır. Aşağıdaki örnek için en esnek fonksiyonun en sağdakidir çünkü verinin yapısına uyum sağlamıştır. Bu yüzden overfitting durumuna ulaşmıştır.
* **Yanlılık**, gerçek değerler ile tahmin edilen değerler arasındaki farkı ifade eder. Aşağıdaki örnek için en yanlı olan en soldaki örnektir çünkü verinin tüm noktalarını temsil edememektedir. Bu yüzden underfitting durumundadır. Bazı gözlemleri çok iyi temsil edebilirken bazı gözlemleri temsil edememektedir, bu sebeple yanlılığı da kafamızda canlandırabiliriz.
* **Varyans**, değişkenlik demektir. Varyans burada modelim esnekliği,hassaslığı belirtmektedir. Yani daha önce belirttiğimiz esneklik aslında varyanstır. Varyansın yüksek olması verisetindeki içerisindeki yapıyı esnek bir şekilde yani temsil gücü yüksek bir şekilde, oldukça detaylı bir şekilde temsil etmek demektir.
* Overfitting durumunda, elindeki örnek verisetini çok iyi öğrenmiştir. Fakat görmediği değerlerle çalışılmak istenildiğinde başarısız olur çünkü elindeki veriyi ezberlemiştir.
* Yanlılık ile varyans arasında ters orantı vardır. Yanlılık bazı verileri iyi temsil etmekken varyans tüm verileri temsil etmektir. **Bu duruma bias-variance tradeoff denir.**
* Underfitting durumunda yanlılık söz konusudur. Örnek veriseti içerisindeki her veriyi temsil edemez.
* Ulaşmak istediğimiz model ise *düşük yanlılık ve düşük varyans*a sahip olmalıdır. Düşük yanlılık ile veriseti içerisindeki bütün verileri temsil edebilme yetisi olsun, düşük varyans ile de gözlemlenmemiş verileri daha iyi temsil edebilsin isteriz.
* *Esneklik neye göre seçilir?* 
    * Ortalama Test Hatası'na göre seçilir. OTH'sı yüksek ise daha esnek bir model seçeriz. Daha esnek bir model seçildiğinde varyans artar ve yanlılık azalır.
* <div> </div> 
* <div> </div> 
* **Learning Curve:**
    * Modelimiziin eğitimini gerçekleştirirken, eğitimde nerede durmamızı gösteren grafik learning curve grafiğidir.
    * Burada model karmaşıklığı, parametrelerin ve hiperparametrelerin artmasıdır.
    * Model karmaşıklığı arttıkça eğitim seti üzerinde tahmin hatası azalmaktadır çünkü model eğitim setini öğrenmeye başlar.
    * Test seti üzerindeki tahmin hatası da bir yere kadar eğitim setiyle birlikte azalmaya başlar çünkü model iyi eğitilmeye devam edilmektedir.
    * Fakat optimum noktadan sonra model eğitim verisini ezberlemeye başladığı için test verisi üzerindeki tahmin hatası da artmaya başlar.
    * Optimum noktaya değişken sayısı, hiperparametre sayısı veya hiperparametrelerin değeri şeklinde isimlendirilebilir.
    * <div> </div>

### Basit Dogrusal Regresyon Modeli Olusturma
Oncelikle verisetimizin hikayesine odaklanalim;
* Veri setimizde 4 adet degisken bulunmaktadir (Unnamed,TV,Radio, Newspaper, Sales)
* Bu degiskenlerden Unnamed isimli olani index olarak kullanilmak icin eklendiginden gozardi edilebilir.
* TV, Radio ve Newspaper degiskeni altinda yer alan  degerler ise bu alanlara harcanmis olan reklam degerleridir.
* Sales altinda yer alan degerler ise reklam harcamalarinin sonucunda olusan satis degerleridir.
* Burada basit dogrusal regresyonu konu aldigimiz icin bir adet bagimsiz degisken ve bagimli degisken secmemiz gerekiyor. Bunun icin bagimsiz olarak TV, bagimli olarak elbette Sales degiskenini seciyoruz.

```python
import pandas as pd
df = pd.read_csv("./csv/Advertising.csv")
df.head()
```

Modellemeyi gerceklestirmeden once verisetinin durumunu gozlemlemek icin **seaborn** kutuphanesi icerisindeki **jointplot** grafigi ile verisetinin durumuna goz atiyoruz.

```python
import seaborn as sns
sns.jointplot(x="TV",y="Sales",data=df, kind="reg");
```

Takibi kolaylastirmak adina X ve Y degisken adlarina gerekli degerlerimizi atalim.

```python
X = df[["TV"]]
Y = df[["Sales"]]
```

* Basit dogrusal regresyon modelini *from sklearn.linear_model import LinearRegression* ifadesiyle projemize dahil ediyoruz. 
* Modelin constructorini cagirarak reg degiskeni icerisinde baslatiyoruz. 
* reg.fit metodu ile modelimizi egitiyoruz. 
* Egittigimiz bu modeli bir degisken icerisine atarak kullanima devam ediyoruz.

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
model = reg.fit(X,Y)
```

* Model degiskeni ozelligi olan
    * intercept_ ozelligi $b_0$ sabitini ifade eder.
    * coef_ ozelligi ise $b_1$ parametresini ifade eder.

```python
print(model.intercept_)
model.coef_
```

$R^2$ Score : Bagimli degiskendeki degisikligin, bagimsiz degiskenlerce aciklanma yuzdesidir.
* Bu ornekte satislardaki degisimin %61'ini TV bagimsiz degiskeni tarafindan aciklanabilmektedir.

```python
model.score(X,Y)
```

Model uzerinde gerceklestirecegimiz tahminler **predict** metodu araciligiyla yapariz. Veriyi verirsek bir liste halinde vermeliyiz, listenin icerisindeki elemanlari da bir liste icerisine yerlestirerek metoda veririz.

```python
model.predict([[165]])
```

Birden fazla deger girmek istersek asagidaki gibi gerceklestirebiliriz.

```python
model.predict([[142],[458],[258]])
```

### Artıklar ve Makine Öğrenmesindeki Önemi
Modelleme ve tahmin algoritmalarını öğrendikten sonra, model optimizasyonu için önemli bir konudur. XGBoost gibi algoritmalar artıklara yönelik optimizasyon gerçekleştirir.
* Veri içerisinden parametreleri öğrendik ve bu parametrelere göre de tahmin değerlerimizi oluşturduk. Şimdi elimizde hem gerçek hem de tahmin edilen değerler bulunmaktadır.
* Amaç, gerçek değerler ile tahmin edilen değerlerin arasındaki farkları minimuma çekmektir. 
* Kurmuş olduğumuz modellerin başarısını değerlendirmek için göz önünde bulunduracağımız metrikler de buna yönelik metriklerdir.
* Regresyon modelleri için *Hata Kareler Ortalaması(MSE)* ve *Hata Kareler Ortalamasının Karekökü(RMSE* metriklerini kullanacağız. 
* Sınıflandırma problemlerinde ise *Doğru Sınıflandırma Oranı*na bakacağız.
